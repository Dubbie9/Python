```python
# The “carb” variable has been found to have incorrect data and is no longer needed. Remove this variable from the data frame.
# To remove data from a dataframe, first preview the data
mtcars = pd.read_csv('/content/mtcars.csv')
print(mtcars)

mtcars.drop(columns = 'carb')
```

```python
#The owner of the garage wants to do some analysis on only the cars with 8-cylinder engines. 
#Create a subset data frame of these cars.
subset = mtcars[mtcars["cyl"] >= 8]
print(subset)
```

```python
#Using the full data (not the subset), you want to explore the relationship between engine size and engine power. 
#Produce a plot with displacement (“disp”) on the x axis and horsepower (“hp”) on the y axis. Use colour to illustrate the quantity of cylinders in the engine
import pandas as pd
import matplotlib.pyplot as plt

#Scatterplot horsepower against displacement
plt.scatter(mtcars['hp'], mtcars['disp'], c=mtcars['cyl'])

#setting x nd y labels
plt.xlabel('displacement')
plt.ylabel('horsepower')

#adding title
plt.title('Relationship between engine size and power')

plt.colorbar()

plt.show()

```
